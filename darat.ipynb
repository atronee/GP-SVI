```python
import pandas as pd
from matplotlib import pyplot as plt
# use the protein datas
```

```python
treino_data = pd.read_csv('200_indu.csv')
plt.figure(figsize=(10,6))
plt.plot(treino_data['epochs'], treino_data['batch_likelihood'], 'r')
plt.xlabel('Epoch')
plt.ylabel('Likelihood')
plt.title('Protein data com 100 variáveis de indução')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(treino_data['epochs'], treino_data['Noise Variance'], 'b')
plt.xlabel('Epoch')
plt.ylabel('Noise Variance')
plt.title('Protein data')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(treino_data['epochs'], treino_data['RBF Lengthscale'], 'y')
plt.xlabel('Epoch')
plt.ylim(0.1, 0.5)
plt.ylabel('RBF Lengthscale')
plt.title('Protein data')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(treino_data['epochs'], treino_data['RBF Variance'], 'g')
plt.xlabel('Epoch')
plt.ylabel('RBF Variance')
plt.title('Protein data')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(treino_data['epochs'], treino_data['mnll'], 'orange')
plt.xlabel('Epoch')
plt.ylabel('Mean negative log likelihood')
plt.title('Protein data')
plt.show()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 8))
fig.suptitle('Protein data com 200 variáveis de indução')

epochs = treino_data['epochs']
batch_likelihood = treino_data['batch_likelihood']
noise_variance = treino_data['Noise Variance']
rbf_lengthscale = treino_data['RBF Lengthscale']
rbf_variance = treino_data['RBF Variance']


# Plot batch_likelihood
axs[0, 0].plot(epochs, batch_likelihood, 'r-', label='batch_likelihood')
axs[0, 0].set_xlabel('Epoch')
axs[0, 0].set_ylabel('Likelihood')
axs[0, 0].legend()

# Plot noise_variance
axs[0, 1].plot(epochs, noise_variance, 'b-', label='Noise Variance')
axs[0, 1].set_xlabel('Epoch')
axs[0, 1].set_ylabel('Noise Variance')
axs[0, 1].legend()

# Plot rbf_lengthscale
axs[1, 0].plot(epochs, rbf_lengthscale, 'y-', label='RBF Lengthscale')
axs[1, 0].set_xlabel('Epoch')
axs[1, 0].set_ylabel('RBF Lengthscale')
# limit the y-axis
axs[1, 0].set_ylim(0.17, 0.3)
axs[1, 0].legend()

# Plot rbf_variance
axs[1, 1].plot(epochs, rbf_variance, 'g-', label='RBF Variance')
axs[1, 1].set_xlabel('Epoch')
axs[1, 1].set_ylabel('RBF Variance')
axs[1, 1].legend()

# Plot mnll
mnll = treino_data['mnll']
axs[2, 0].plot(epochs, mnll, 'orange', label='Mean negative log likelihood')
axs[2, 0].set_xlabel('Epoch')
axs[2, 0].set_ylabel('Mean negative log likelihood')
axs[2, 0].legend()

# delete the last subplot
fig.delaxes(axs[2,1])


# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show plot
plt.show()
```
